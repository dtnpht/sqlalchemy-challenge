```python
import pandas as pd
from datetime import datetime as dt
```

```python
# "tobs" is "temperature observations"
df = pd.read_csv('./Resources/hawaii_measurements.csv')
df.head()
```

```python
# Convert the date column format from string to datetime
df['date'] = pd.to_datetime(df['date'])
```

```python
# Set the date column as the DataFrame index
df2 = df.set_index(df['date'])
```

```python
# Drop the date column
df2.drop('date',axis=1, inplace=True)
```

```python
df2
```

### Compare June and December data across all years 

```python
from scipy import stats
```

```python
# Filter data for desired months
june_df = df2.filter(like='-06-', axis=0)
june_df
```

```python
dec_df = df2.filter(like='-12-',axis=0)
dec_df
```

```python
# Identify the average temperature for June
temp_avg_june = june_df.groupby('date').mean()
temp_avg_june
```

```python
# Identify the average temperature for December
temp_avg_dec = dec_df.groupby('date').mean()
temp_avg_dec
```

```python
# Create collections of temperature data
```

```python
# Run paired t-test
stats.ttest_ind(temp_avg_june['tobs'], temp_avg_dec['tobs'])
```

### Analysis

Since both months data has their own significants, using a paired t-test cannot show the relationship to each other. But we can do unpaired t-test for each month then doing comparison between two results, which give data anallytics some visualization about the outcomes from two months.
